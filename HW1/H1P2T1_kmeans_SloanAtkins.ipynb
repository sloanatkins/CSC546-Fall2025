#### Handle the categorical variable "ocean_proximity"
Many machine learning algorithms only accept numerical data (numbers) not categorical data (category names). Thus, we need to convert categorical data into numerical data. <br>
In our table, we have a categorical variable "ocean_proximity". We apply one-hot encoding method (https://en.wikipedia.org/wiki/One-hot) to use numbers to represent different "ocean_proximity", which can be easily done by Pandas.

```python
#get a new table by selecting columns related to location
#assume the name of the new table is data1
#for example:
#    data1 = housing[['longitude', 'latitude']] to select only two columns
#    data1.head() to show the first five rows

# Select location-related columns (lon/lat + one-hot ocean_proximity)
# Minimal location table with just longitude/latitude
# Clean column names just in case

housing.columns = housing.columns.str.strip()
base = housing[['longitude', 'latitude']].copy()

if 'ocean_proximity' in housing.columns:
    dummies = pd.get_dummies(housing['ocean_proximity'],
                             prefix='ocean_proximity', dtype='int8')
else:
    dummy_cols = [c for c in housing.columns if c.startswith('ocean_proximity_')]
    dummies = housing[dummy_cols].astype('int8')

data1 = pd.concat([base, dummies], axis=1)
data1.head()

```

```python
#type of data1 is DataFrame
#convert DataFrame to numpy array

X = data1.to_numpy(dtype=float)
```

```python
#normalize data1 using MinMaxScaler in sk-learn
#assume that data1n is the name of the normalized data1
#    data1 -> normlaization -> data1n
#use fit_transform of MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data1n = scaler.fit_transform(X)
```

```python
data1n.shape
```

```python
#run k-means on data1n
#set n_clusters to 3
#read the docmument of k-means on sk-learn website
#assume the cluster labels are stored in the array label1

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(data1n)
label1 = kmeans.labels_
```

```python
#show the clusters using different colors
#you can choose a color in ['r', 'g', 'b', 'm', 'c', 'y']
#code to get the data points in cluster-k
#  cluster_k=data1[label1==k]
#assume 'longitude' is in column-0 of data1
#code to get longitude values of the data points in cluster-k
#  longitude_k=data1[label1==k, 0]
# sample code is here, assuming three clusters
# color=['r', 'g', 'b', 'm', 'c', 'y']
# fig, ax = plt.subplots(1,3, figsize=(10,6))
# for k in range(0, 3):
#    ax[k].imshow(california_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.5)
#    ax[k].scatter(x=data1[label1==k,0], y=data1[label1==k,1],
#                  color=color[k],  marker='.', s=1, alpha=0.4)
#    ax[k].set_ylabel("Latitude", fontsize=14)
#    ax[k].set_xlabel("Longitude", fontsize=14)

import matplotlib.pyplot as plt

california_img = plt.imread("/content/california.png")

coords = data1[['longitude', 'latitude']].to_numpy()

colors = ['r', 'g', 'b']
fig, ax = plt.subplots(1, 3, figsize=(12, 4), sharex=True, sharey=True)

for k in range(3):
    ax[k].imshow(california_img, extent=[-124.55, -113.80, 32.45, 42.05], alpha=0.5)
    ax[k].scatter(
        coords[label1 == k, 0],  # longitude
        coords[label1 == k, 1],  # latitude
        c=colors[k], marker='.', s=1, alpha=0.4
    )
    ax[k].set_ylabel("Latitude", fontsize=14)
    ax[k].set_xlabel("Longitude", fontsize=14)

plt.tight_layout()
```

```python
#plot the histogram of cluster labels in label1
import numpy as np
import matplotlib.pyplot as plt

K = kmeans.n_clusters
vals, counts = np.unique(label1, return_counts=True)

plt.figure(figsize=(6,4))
plt.bar(vals, counts)
plt.xticks(range(K))
plt.xlabel("Cluster label")
plt.ylabel("Count")
plt.title("Cluster sizes")
plt.show()
```

```python
#run k-means on data2n to get cluster labels
#you can try to set n_clusters to 6
#assume the cluster labels are stored in the array label2

from sklearn.cluster import KMeans
kmeans2 = KMeans(n_clusters=6, n_init=10, random_state=42)
label2 = kmeans2.fit_predict(data2n)
```

```python
#show the clusters using different colors
#see cell 24

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

cent_unscaled = scaler2.inverse_transform(kmeans2.cluster_centers_)

order = np.argsort(cent_unscaled[:, 2])

coords2 = data2[['longitude','latitude']].to_numpy()
extent  = [-124.55, -113.80, 32.45, 42.05]
colors  = ['r','g','b','m','c','y']

california_img = None
for p in [Path("california.png"), Path("/content/california.png"), Path("/mnt/data/california.png")]:
    if p.exists():
        california_img = plt.imread(str(p))
        break

fig, ax = plt.subplots(1, 6, figsize=(12, 2.6), sharex=True, sharey=True)
for panel_idx, orig_label in enumerate(order):
    if california_img is not None:
        ax[panel_idx].imshow(california_img, extent=extent, alpha=0.5, zorder=0)

    mask = (kmeans2.labels_ == orig_label)
    ax[panel_idx].scatter(coords2[mask, 0], coords2[mask, 1],
                          c=colors[panel_idx], marker='.', s=2, alpha=0.6, zorder=1)
    ax[panel_idx].set_xlim(extent[0], extent[1])
    ax[panel_idx].set_ylim(extent[2], extent[3])
    ax[panel_idx].set_xticks([]); ax[panel_idx].set_yticks([])
    ax[panel_idx].set_aspect('equal', adjustable='box')
    ax[panel_idx].set_title(f"Cluster {panel_idx}", fontsize=10)

plt.tight_layout(w_pad=0.3)
plt.show()
```
